```python
from ultralytics import YOLO
import os
import matplotlib.pyplot as plt
import cv2
```

```python
model = YOLO("yolov8n.pt")  # lightweight model (good for student project)
```

```python
import os

print(os.path.exists(r"..\dataset\NEU-DET-YOLO\data.yaml"))
```

```python
import yaml

with open(r"..\dataset\NEU-DET-YOLO\data.yaml", "r") as f:
    data = yaml.safe_load(f)

print(data)
```

```python
results = model.train(
    data=r"..\dataset\NEU-DET-YOLO\data.yaml",
    epochs=10,
    imgsz=640,
    batch=2,
    lr0=0.001,
    device="cpu"
)
```

```python
metrics = model.val()

print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
print("mAP@0.5:", metrics.box.map50)
print("mAP@0.5:0.95:", metrics.box.map)
```
